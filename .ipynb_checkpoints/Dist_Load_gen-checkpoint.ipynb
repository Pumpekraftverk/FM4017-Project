```python
import pypsa
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
from fetch_entose import fetch_entsoe_data

# Path to your solved network
network_path = "base_s_70_elec_.nc"

# Load it
n = pypsa.Network(network_path)

# Remove elecotrlysis and batteries buses
targets = {"h2", "battery"}
buses = n.buses.index[n.buses.carrier.astype(str).str.casefold().isin(targets)]
n.remove("Bus", buses)

print(n.buses['carrier'].value_counts())
```

```python
# Load the bidding zones GeoJSON
zones_gdf = gpd.read_file("bidding_zones.geojson")

# Compute bus loading (sum over all time steps)
bus_loading = n.loads_t.p.sum(axis=0)  # MW per bus

# Extract line capacities
line_capacities = n.lines.s_nom.copy()
line_widths = 1 #+ 8 * line_capacities / line_capacities.max()

# Extract link capacities
if not n.links.empty:
    link_capacities = n.links.p_nom.copy()
    link_widths = 1 #+ 4 * link_capacities / link_capacities.max()
else:
    link_widths = None

# Create GeoAxes
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection=ccrs.PlateCarree())

# --- Plot bidding zones (as polygons) ---
zones_gdf.plot(ax=ax, edgecolor="black", facecolor="none", linewidth=1)

# If you want filled colors per zone:
#zones_gdf.plot(ax=ax, column="name", legend=True, alpha=0.3)

# --- Plot the network ---
n.plot.map(ax=ax,
           bus_colors=bus_loading,
           bus_sizes=0.01,
           line_alpha=1,
           line_widths=line_widths,
           link_alpha=1,
           link_widths=link_widths)

# Add colorbar for bus loadings
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis,
                           norm=plt.Normalize(vmin=bus_loading.min(),
                                              vmax=bus_loading.max()))
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.7)
cbar.set_label("Total Load [MW]")

plt.title("PyPSA Network with Bidding Zones, Bus Loading, Line & Link Capacities")
plt.show()
```

```python
#rename map
fix = {
    # Norway
    "NOS1":"NO_1","NOS2":"NO_2","NOM1":"NO_3","NON1":"NO_4","NOS5":"NO_5",
    # Sweden
    "SE01":"SE_1","SE02":"SE_2","SE03":"SE_3","SE04":"SE_4",
    # Finland
    "FI00":"FI",
    # Denmark
    "DKE1":"DK_2",    # east DK (ENTSO-E DKE1)  DK_2
    "DKW1":"DK_1",    # west DK (you may want to keep/drop this; see below)
    # Neighbours (so flows/buses dont look alien)
    "NL00":"NL",
    "DE00":"DE_LU",
    "EE00":"EE",
    "LT00":"LT",
    "PL00":"PL",
    "GB00":"GB",
    "GBNI":"GB",
}

# Buses as points (same CRS as zones)
bus_gdf = gpd.GeoDataFrame(
    n.buses,
    geometry=gpd.points_from_xy(n.buses.x, n.buses.y),
    crs=zones_gdf.crs,
)
# Join and assign
j = gpd.sjoin(bus_gdf, zones_gdf[["zone_name","geometry"]], how="left", predicate="within")
n.buses["zone"] = j["zone_name"].replace(fix)
n.generators['zone'] = n.generators["bus"].map(n.buses["zone"])

n.buses.loc['SE1 6', 'zone'] = 'SE_3'
n.buses.loc['DK1 0', "zone"] = 'DK_2'
print(n.buses['zone'].value_counts(dropna=False))
print(n.generators['zone'].value_counts(dropna=False))
```

```python
loads_df, gens_df, flows_df = fetch_entsoe_data(start_date="2025-09-01", timezone="Europe/Oslo")
```

```python
n.set_snapshots(loads_df.index)
zone_to_buses = {z: list(idx) for z, idx in n.buses.groupby("zone").groups.items()}
print(gens_df.head())
```

```python
for z in loads_df.columns:
    buses = zone_to_buses.get(z, [])
    if not buses:
        raise ValueError(f"No buses found in zone {z}")
    # each bus must already have a Load with the SAME name as the bus
    miss = [b for b in buses if b not in n.loads.index]
    if miss:
        raise ValueError(f"Missing loads (load name must equal bus name) for buses: {miss}")

    share = loads_df[z].reindex(n.snapshots).fillna(0) / len(buses)
    for b in buses:
        n.loads_t.p_set[b] = share
```

```python
# 2) map generators to zones (via their bus)
tech = n.generators.index.to_series().str.rsplit(n=1).str[-1].str.lower()
tech.unique()
```

```python
map_tech = {
    "ccgt": "Fossil Gas",
    "ocgt": "Fossil Gas",
    "biomass": "Biomass",
    "coal": "Fossil Hard coal",
    "lignite": "Fossil Hard coal", 
    "oil": "Fossil Oil",
    "nuclear": "Nuclear",
    "onwind": "Wind Onshore",
    "offwind-ac": "Wind Offshore",
    "offwind-dc": "Wind Offshore",
    "offwind-float": "Wind Offshore",
    "solar": "Solar",
    "solar-hsat": "Solar",
    "ror": "Hydro Run-of-river and poundage",
}

gen_tech = tech.map(map_tech)
```

```python
p_set = pd.DataFrame(0.0, index=n.snapshots, columns=n.generators.index)

for (T, Z), gens in n.generators.index.to_series().groupby([gen_tech, n.generators["zone"]]):
    if (T, Z) not in gens_df.columns: 
        continue
    s = gens_df[(T, Z)].to_numpy()[:, None]
    w = n.generators.loc[gens, "p_nom"].to_numpy()
    w = w / w.sum() if w.sum() > 0 else np.full(len(gens), 1/len(gens))
    p_set[gens] = s * w

n.generators_t.p_set = p_set
```
